```python
import numpy as np
import matplotlib.pyplot as plt

f = 60
t = np.linspace(0,1,1000)
x = np.sin(2*np.pi*5*t)+np.sin(2*np.pi*f*t)

x2 = np.sin(2*np.pi*f*t)
convol=np.convolve(x,x2)
```

```python
plt.figure(figsize=(16,3))
plt.subplot(1,2,1)
plt.plot(x)
plt.subplot(1,2,2)
plt.plot(x2)
plt.show()

plt.figure(figsize=(15.78,3))
plt.plot(convol)
plt.show()
```
